```python
from pandas import to_datetime, date_range
import pandas as pd
from datetime import datetime


def export_tables_side_by_side(dfs, base_path_without_ext, sheet_name="Data"):
    from openpyxl import Workbook, load_workbook
    from openpyxl.worksheet.table import Table, TableStyleInfo

    file_path = f"{base_path_without_ext}.xlsx"
    print(f"Writing XLSX: {file_path}")

    # fresh workbook
    wb = Workbook()
    ws = wb.active
    ws.title = sheet_name
    wb.save(file_path)

    # reload for writing
    wb = load_workbook(file_path)
    ws = wb[sheet_name]

    # ----------------------------
    # write headers + data using openpyxl ONLY
    # ----------------------------
    start_col = 1
    for df in dfs:
        # write headers
        for col_idx, col_name in enumerate(df.columns, start=start_col):
            ws.cell(row=1, column=col_idx, value=str(col_name))

        # write data rows
        for row_idx, row in enumerate(df.itertuples(index=False), start=2):
            for col_offset, cell_val in enumerate(row):
                ws.cell(
                    row=row_idx,
                    column=start_col + col_offset,
                    value=cell_val
                )

        start_col += len(df.columns) + 1  # spacer

    # ----------------------------
    # add tables
    # ----------------------------
    start_col = 1
    for i, df in enumerate(dfs):
        end_col = start_col + len(df.columns) - 1
        end_row = len(df) + 1

        start_letter = ws.cell(row=1, column=start_col).column_letter
        end_letter   = ws.cell(row=1, column=end_col).column_letter
        table_ref = f"{start_letter}1:{end_letter}{end_row}"

        table_name = f"Table_{i}"
        t = Table(displayName=table_name, ref=table_ref)
        t.tableStyleInfo = TableStyleInfo(
            name="TableStyleMedium9",
            showFirstColumn=False,
            showLastColumn=False,
            showRowStripes=True,
            showColumnStripes=False
        )
        ws.add_table(t)

        start_col += len(df.columns) + 1

    wb.save(file_path)
    print(f"   XLSX Done: {file_path}")


# =============================
# main
# =============================

start = to_datetime("2026-01-05")
end   = to_datetime("2026-04-01")
modelPath = r"exp\keras\keras-1\2026_01_01_01_43_15\e32__l128-64-64__ep20__oa_linear__134912"
dfs = []

for d1 in date_range(start, end, freq="1D"):
    pred_df = groceryML.RunPredictionsOnly(groceryML.combined_df, modelPath, prediction_date=d1)

    block = pred_df[["date", "item", "prediction"]] \
            .sort_values("prediction", ascending=False) \
            .head(50) \
            .reset_index(drop=True)

    block.columns = [str(c) for c in block.columns]
    dfs.append(block)

ts = datetime.now().strftime("%Y_%m_%d_%H_%M_%S")
outPath = os.path.join(modelPath, f"predictions_side_by_side-{ts}")
print (f"Outpath: {outPath}");
       
export_tables_side_by_side(dfs, outPath, "preds")

print("done")
```
