```python
from tensorflow import keras
import pandas as pd
import shap
import numpy as np

# --- paths you edit ---
model_path = r"C:\Users\steve\source\repos\grocery-ml\exp\keras\2025_12_25_19_45_10\e128__l1024-512-32__ep15__oa_linear__213572\model"          # folder containing saved_model.pb
data_path  = r"C:\Users\steve\source\repos\grocery-ml\exp\keras\2025_12_25_19_45_10\e128__l1024-512-32__ep15__oa_linear__213572\predictions-e128__l1024-512-32__ep15__oa_linear__213572.csv"     # same columns as training input

# --- load model and data ---
model = keras.models.load_model(model_path)
df = pd.read_csv(data_path)

# --- split into two input tensors ---
X_num = df.drop(columns=["itemId"]).values       # numeric block
X_id  = df["itemId"].values.reshape(-1,1)        # id/index block

# --- build masker from numeric input only ---
masker = shap.maskers.Independent(X_num[:100])

# --- explainer for a multi-input Keras model ---
explainer = shap.Explainer(model, masker)

# --- build a matching sample input list ---
sample_num = X_num[:200]
sample_id  = X_id[:200]

# --- shap values ---
shap_values = explainer([sample_num, sample_id])

# --- visualize numeric feature effects only ---
shap.plots.beeswarm(shap_values[0])
```
